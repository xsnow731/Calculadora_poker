```python
# Instalar biblioteca
!pip install treys

import random
from treys import Card, Deck, Evaluator

def simular_equidade(cartas_jogador, n_oponentes=1, n_simulacoes=100000):
    avaliador = Evaluator()
    mao_jogador = [Card.new(c) for c in cartas_jogador]

    vitorias = 0
    empates = 0

    for _ in range(n_simulacoes):
        deck = Deck()
        for c in mao_jogador:
            deck.cards.remove(c)

        # Oponentes
        maos_oponentes = []
        for _ in range(n_oponentes):
            maos_oponentes.append([deck.draw(1)[0], deck.draw(1)[0]])

        # Mesa
        board = deck.draw(5)

        # Jogador
        score_jogador = avaliador.evaluate(board, mao_jogador)

        # Oponentes
        scores_oponentes = [avaliador.evaluate(board, mao) for mao in maos_oponentes]

        menor_score = min([score_jogador] + scores_oponentes)

        if score_jogador == menor_score:
            if scores_oponentes.count(score_jogador) > 0:
                empates += 1
            else:
                vitorias += 1

    prob_vitoria = vitorias / n_simulacoes
    prob_empate = empates / n_simulacoes
    prob_derrota = 1 - (prob_vitoria + prob_empate)

    return {
        "Vitória": round(prob_vitoria * 100, 2),
        "Empate": round(prob_empate * 100, 2),
        "Derrota": round(prob_derrota * 100, 2)
    }

# ================================
# Entrada do usuário
# ================================

print("Digite suas duas cartas no formato: ValorNaipe")
print("Exemplo: As (Ás de espadas), Kd (Rei de ouros), Th (10 de copas), 7c (7 de paus)")

carta1 = input("Digite a primeira carta: ").strip()
carta2 = input("Digite a segunda carta: ").strip()
n_oponentes = int(input("Digite o número de adversários: "))

resultado = simular_equidade([carta1, carta2], n_oponentes=n_oponentes, n_simulacoes=50000)
print("\nProbabilidade aproximada em %:")
for k, v in resultado.items():
    print(f"{k}: {v}%")
```
